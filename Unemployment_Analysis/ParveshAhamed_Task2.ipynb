```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
plt.style.use('ggplot')
```

```python
df = pd.read_csv("C:/Users/parvesh ahamed/Downloads/Unemployment in India task 2.csv")
print(df.head())
print(df.shape)
```

```python
print(df.info())
print(df.describe())
```

```python
print(df.isnull().sum())
```

```python
df.columns = df.columns.str.strip()
print(df.columns)
```

```python
df = df.dropna()
print(df.isnull().sum())
```

```python
print(df.columns)
```

```python
df.columns = df.columns.str.strip()
print(df.columns)
```

```python
plt.figure(figsize=(8,5))
sns.histplot(df['Estimated Unemployment Rate (%)'], bins=20)
plt.title("Distribution of Unemployment Rate")
plt.show()
```

```python
region_unemployment = df.groupby('Region')['Estimated Unemployment Rate (%)'].mean()
region_unemployment = region_unemployment.sort_values(ascending=False)
plt.figure(figsize=(12,6))
region_unemployment.plot(kind='bar')
plt.title("Average Unemployment Rate by Region")
plt.ylabel("Unemployment Rate (%)")
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.scatterplot(
    x='Estimated Labour Participation Rate (%)',
    y='Estimated Unemployment Rate (%)',
    data=df)
plt.title("Labour Participation vs Unemployment")
plt.show()
```

```python
plt.figure(figsize=(7,5))
sns.boxplot(
    x='Area',
    y='Estimated Unemployment Rate (%)',
    data=df)
plt.title("Urban vs Rural Unemployment")
plt.show()
```

```python
numeric_df = df.select_dtypes(include=np.number)
plt.figure(figsize=(8,6))
sns.heatmap(
    numeric_df.corr(),
    annot=True,
    cmap='coolwarm')
plt.title("Correlation Heatmap")
plt.show()
```
